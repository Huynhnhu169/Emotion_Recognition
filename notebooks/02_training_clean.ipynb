# Leakage-Safe Training

This notebook delegates training to `src.train`. The pipeline splits original files before augmentation, then augments only the training split.

```python
from pathlib import Path
import os
import sys

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
os.chdir(ROOT)
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

CONFIG_PATH = ROOT / 'configs' / 'emotion_model.yaml'

from src.data_split import load_config, load_or_create_splits

config = load_config(CONFIG_PATH)
splits = load_or_create_splits(config, save=True)
for name, frame in splits.items():
    print(name, len(frame), frame['label'].value_counts().to_dict())
```
